```python
import os
import pandas as pd
import json
import tqdm
import numpy as np
import re
```

```python
records = os.listdir("MULTI_RUN_TEMP/")
records = [_ for _ in records if _.split(".")[-1]=='json']
records = sorted(records)
records
```

```python
len(records)
```

```python
df_raw = pd.DataFrame()
```

```python
for record in tqdm.tqdm(records):
    # To read the dictionary from the JSON file
    with open(f'MULTI_RUN_TEMP/{record}', 'r') as json_file:
        loaded_record = json.load(json_file)
    
    #if loaded_record['model'] in ['AutoStreamML','OnlineAutoML','EvoAutoML','Chcha']:
    rec = {
        "run":record.split("_")[-1].split('.')[0],
            "model":loaded_record['model'],
           "dataset":loaded_record['dataset'],
           "r2_score":loaded_record['regressionEvaluator']['coefficient of determination'],
        "rmse":loaded_record['regressionEvaluator']['root mean squared error'],
          "time":abs(sum([i for i in loaded_record['time']])),
           "memory":abs(sum([i/1024/1024  for i in loaded_record['memory']]))
          }
    df_raw = df_raw._append(rec,ignore_index=True)
```

```python
df = df_raw.copy()
```

# Creating multi-run table

```python
temp_df = df.groupby(['model','dataset'])['r2_score'].aggregate(['mean','std'])
temp_df.reset_index(inplace=True)
```

```python
temp_df
```

```python
temp_df['score'] = [f"{mean:.2f}±{std:.2f}" for mean, std in zip(temp_df['mean'] * 100, temp_df['std'] * 100)]
```

```python
temp_df = temp_df.pivot_table(index='dataset',columns='model',values='score',aggfunc='first').reset_index(drop=False).reset_index(drop=True)
temp_df.columns.name=None
temp_df = temp_df.set_index('dataset')
temp_df = temp_df.reindex(dataset_name_list)
temp_df.index = [" ".join(i.split("_")).title() for i in temp_df.index]
temp_df = temp_df[columns_order]
temp_df.columns = columns_rename
temp_df
```

```python
temp_df.to_clipboard()
```

```python
def add_mean_row_to_dataframe(_df):
    # Calculate the mean for each column and create a DataFrame for the mean values
    mean_values = _df.mean().to_frame().transpose()
    
    # Append the mean values DataFrame to the original DataFrame
    df_with_mean = _df._append(mean_values, ignore_index=False)
    
    df_with_mean.rename({df_with_mean.index[-1]: 'Mean'}, inplace=True)
    
    return df_with_mean

def add_mean_rank_row_to_dataframe(_df,ascending=True):
    
    df_temp = _df[:-1].rank(axis=1,ascending=ascending)
    
    # Calculate the mean for each column and create a DataFrame for the mean values
    mean_values = df_temp.mean().to_frame().transpose()
    
    # Append the mean values DataFrame to the original DataFrame
    df_with_mean = _df._append(mean_values, ignore_index=False)
    
    df_with_mean.rename({df_with_mean.index[-1]: 'Mean_Rank'}, inplace=True)
    
    return df_with_mean
```

```python
if 'Mean_Rank' not in dataset_name_list:
    dataset_name_list +=['Mean','Mean_Rank']
    #dataset_name_list +=['Mean_Rank']
    print("Added Mean and Mean Rank")
```

```python
temp_df = df.pivot_table(index='dataset',columns='model',values='r2_score',aggfunc='mean').reset_index(drop=False).reset_index(drop=True)
#temp_df = df.pivot(index='dataset',columns='model',values='r2_score').reset_index(drop=False).reset_index(drop=True)
temp_df.columns.name=None
temp_df = temp_df.set_index('dataset')#.multiply(100)
temp_df = temp_df[columns_order]
temp_df = add_mean_row_to_dataframe(temp_df).multiply(100)
temp_df = add_mean_rank_row_to_dataframe(temp_df,ascending=False)
temp_df = temp_df.reindex(dataset_name_list)
temp_df.index = [" ".join(i.split("_")).title() for i in temp_df.index]
temp_df.columns = columns_rename
temp_df.style.highlight_max(color = 'yellow', axis = 1)
```
